```sql
USE bronze;
```

```sql
SHOW TABLES IN bronze;
```

```sql
CREATE TABLE bronze.events
USING DELTA
AS SELECT * FROM delta.`/Volumes/workspace/ecommerce/ecommerce_data/bronze/events`;
```

```sql
SHOW TABLES IN bronze;
```

```sql
CREATE TABLE silver.events
USING DELTA
AS
SELECT *
FROM bronze.events
WHERE price > 0;
```

```sql
SELECT COUNT(*) FROM silver.events;
```

```sql
CREATE TABLE gold.products AS
SELECT
  product_id,
  COUNT(*) AS total_events,
  SUM(price) AS total_revenue
FROM silver.events
GROUP BY product_id;
```

```sql
SELECT * FROM gold.products LIMIT 10;
```

```sql
CREATE VIEW gold.top_products AS
SELECT
  product_id,
  total_revenue
FROM gold.products
WHERE total_revenue > 1000
ORDER BY total_revenue DESC;
```

```sql
SELECT * FROM gold.top_products LIMIT 10;
```

```sql
DESCRIBE TABLE EXTENDED bronze.events;
DESCRIBE TABLE EXTENDED silver.events;
DESCRIBE TABLE EXTENDED gold.products;
```
