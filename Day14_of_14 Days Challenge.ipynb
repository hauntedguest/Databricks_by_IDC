```python
from transformers import pipeline
from datetime import datetime
import mlflow

def log(msg):
    print(f"[{datetime.now()}] ðŸ”¹ {msg}")

log("Loading sentiment analysis model")

classifier = pipeline("sentiment-analysis")
```

```python
reviews = [
    "This product is amazing!",
    "Terrible quality, waste of money"
]

log("Running sentiment analysis")

results = classifier(reviews)
for review, result in zip(reviews, results):
    print(review, "â†’", result)
```

```python
with mlflow.start_run(run_name="sentiment_model"):
    log("Logging sentiment model experiment")

    mlflow.log_param("model_name", "distilbert-base-uncased-finetuned-sst-2-english")
    mlflow.log_param("task", "sentiment-analysis")

    # Example metric (demo purpose)
    mlflow.log_metric("sample_accuracy", 0.95)

log("MLflow sentiment run completed")
```
