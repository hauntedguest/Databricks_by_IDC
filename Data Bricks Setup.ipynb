```python
def load_ecommerce_data(month):

    path_map = {
        "Oct": "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv",
        "Nov": "/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv"
    }

    if month not in path_map:
        raise ValueError(f"Invalid month: {month}")
    
    df = (
        spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(path_map[month])
    )
    
    return df
```

```python
# Load your data
events = load_ecommerce_data("Oct")

# Verify it's working
print(f"âœ… Ready to go! Loaded {events.count():,} events")
events.show(5, truncate=False)
```
